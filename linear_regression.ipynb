```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
x,y=datasets.make_regression(n_samples=100,n_features=5)
```

```python
plt.scatter(x[:,0],y) # for one feature and target
```

```python
def fit(data,target):
    n_sample,n_feature=data.shape
    
    weights = np.zeros(n_feature)
    #assign weights to zero
    
    bias=0  #assign bias/intercept to zero
    
    l_rate=0.01  #learning rate
    
    for i in range(2000):
        ypred = np.dot(data,weights) + bias   # y' = W.X + b  ;where W=[w1,w2,w3,...,wn] and X=[x1,x2,x3,...,xn]
        
        #cost function = (1/n) sigma{(y- ypred)^2} = (1/n) sigma{(y-( W.X + b ))}
        
        dw = np.dot(data.T,(ypred-target))*(2/n_sample)  #  dw = derivative of cost function wrt W vector 
                                                         #       = (2/n) sigma{(y- ypred)}.W 
        
        db = np.sum(ypred-target)*(2/n_sample)           #  db = derivative of cost function wrt b  
                                                         #      = (2/n) sigma{(y- ypred)} 
        
        weights -= l_rate*dw    # new_wt = old_wt - learning rate * dw ; neagtive sign for opposite direction of steepest 
        
        bias -= l_rate*db    # new_b = old_b - learning rate * db ; neagtive sign for opposite direction of steepest 
    
    return weights,bias

def predict(data,weights,bias):
    return np.dot(data,weights) + bias   # ypred  = W.X + b  ;where W=[w1,w2,w3,...,wn] and X=[x1,x2,x3,...,xn]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
w,b=fit(X_train,y_train)

ypred=predict(X_test,w,b)

r2_score(y_test,ypred)
```
