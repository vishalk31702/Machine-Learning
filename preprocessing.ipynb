```python
# by using inbuilt method ffill or bfill
age1= age.fillna(method='ffill')
print("before : ",age.isnull().sum())
print("after : ",age1.isnull().sum())
age1

#here we had null value at row '888' so ffill will fill this null value by its previous value at '887' which is 19.0
#similarly bfill will fill this null value at '888' with value at '889' which is 26.0
```

so these are the methods to handle null values for any column with dtype= int or float

now we will look for columns having dtype=object means they have string as their data
in this we cant fill them using mean/median. we can use either ffill or bfill. or we can use the most frequent value/mode

we use encodind to convert string into number, there are twotype of categorical data one is 
nonimal(without any order e.g. male/female, yes/no) other is ordinal(with order e.g. worst/good/better, first/second/third)

we have total three techniques : one hot(nonimal) encoding, ordinal encoding, label encoding

label encoding : same as ordinal encoding but according to documentation we only use this for target classes pnly if required
                  although our teacher allow us to apply this to feature also so we can do this in exam also for features

                 then we provide 1 and 0 to them
    
    gender      male    female
    
    male         1       0
    female       0       1
    male         1       0

          best - 2  good - 1  bad - 0
    
    review      review
    
    good        1

    good        1

```python
#2) z-score

def z_score(data):
    outliers=[]
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z=(i-mean)/std
        if abs(z)>3:
            outliers.append(i)
    return outliers

print("outliers in age : ",z_score(df_data["Age"]))
print("outliers in fare : ",z_score(df_data["Fare"]))
```

```python
#3) IQR

def iqr(data):
    outliers=[]
    q1=np.quantile(data,0.25)
    q3=np.quantile(data,0.75)
    iqr=q3-q1
    upper_bound = q3+iqr*(1.5)
    lower_bound = q1-iqr*(1.5)
    
    for i in data:
        if i>upper_bound or i<lower_bound:
            outliers.append(i)
    return outliers   

print("outliers in age : ",iqr(df_data["Age"]))
print("outliers in fare : ",iqr(df_data["Fare"]))
```

#normalization and standarization
used for feature scaling means change the scale of features for better performance

standarization - scalled the data and mean become zero and standard deviation become 1, that means we get centered data
1) z-score

```python
demo=df["Age"]

def z_stand(data):
    mean=np.mean(data)
    std=np.std(data)
    new_value=[]
    for i in data:
        z=(i-mean)/std
        new_value.append(z)
    data = new_value
    return data
z_stand(demo)        
```

normalization

1) minmax

```python
demo=df["Age"]

def min_max(data):
    mini=np.min(data)
    maxi=np.max(data)
    new_value=[]
    for i in data:
        z=(i-mini)/(maxi-mini)
        new_value.append(z)
    data = new_value
    return data
min_max(demo)        
```

```python
from sklearn.preprocessing import MinMaxScaler
std=MinMaxScaler()

scale=std.fit_transform(df_data)
scaled=pd.DataFrame(scale,columns=df_data.columns)
scaled
```

2) normalizer

```python
demo=df["Age"]

def normalizer(data):
    mini=np.min(data)
    maxi=np.max(data)
    mean = np.mean(data)
    new_value=[]
    for i in data:
        z=(i-mean)/(maxi-mini)
        new_value.append(z)
    data = new_value
    return data
normalizer(demo)        
```

```python
from sklearn.preprocessing import Normalizer
std=Normalizer()

scale=std.fit_transform(df_data)
scaled=pd.DataFrame(scale,columns=df_data.columns)
scaled
```

3) Robust

```python
demo=df["Age"]

def robust(data):
    q1=np.quantile(data,0.25)
    q3=np.quantile(data,0.75)
    iqr=q3-q1
    median = np.median(data)
    new_value=[]
    for i in data:
        z=(i-median)/iqr
        new_value.append(z)
    data = new_value
    return data
robust(demo)        
```

```python
from sklearn.preprocessing import RobustScaler
std=RobustScaler()

scale=std.fit_transform(df_data)
scaled=pd.DataFrame(scale,columns=df_data.columns)
scaled
```
